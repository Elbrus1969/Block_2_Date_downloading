# Юнит 4. Детальный анализ вакансий

1. Напишите запрос, который позволит узнать, сколько (cnt) вакансий в каждом регионе (area).
Отсортируйте по количеству вакансий в порядке убывания.

```python
# текст запроса
query_4_1 = f'''select
                    count(v.id) cnt,
                    a.name area
                from vacancies v
                join areas a on v.area_id = a.id
                group by a.name
                order by cnt desc
             '''
with engine.begin() as conn:
    df_4_1 = pd.read_sql_query(sa.text(query_4_1), conn)
```

```python
# результат запроса
df_4_1
```

2. Напишите запрос, чтобы определить у какого количества вакансий заполнено хотя бы одно из двух полей с зарплатой.

```python
# текст запроса
query_4_2 = f'''select
                    count(v.id) cnt
                from vacancies v
                where v.salary_from is not null or v.salary_to is not null
             '''
with engine.begin() as conn:
    df_4_2 = pd.read_sql_query(sa.text(query_4_2), conn)
```

```python
# результат запроса
df_4_2
```

3. Найдите средние значения для нижней и верхней границы зарплатной вилки. Округлите значения до целого.

```python
# текст запроса
query_4_3 = f'''select
                    avg(v.salary_from),
                    avg(v.salary_to)
                from vacancies v
             '''
with engine.begin() as conn:
    df_4_3 = pd.read_sql_query(sa.text(query_4_3), conn)
```

```python
# результат запроса
round(df_4_3).astype('int32')
```

4. Напишите запрос, который выведет количество вакансий для каждого сочетания типа рабочего графика (schedule) и типа трудоустройства (employment), используемого в вакансиях. Результат отсортируйте по убыванию количества.


```python
# текст запроса
query_4_4 = f'''
                select
                    count(v.id) cnt,
                    v.schedule,
                    v.employment
                from vacancies v
                group by v.schedule, v.employment
                order by cnt desc
             '''
with engine.begin() as conn:
    df_4_4 = pd.read_sql_query(sa.text(query_4_4), conn)
```

```python
# результат запроса
df_4_4
```

5. Напишите запрос, выводящий значения поля Требуемый опыт работы (experience) в порядке возрастания количества вакансий, в которых указан данный вариант опыта.

```python
# текст запроса
query_4_5 = f'''
                select
                    count(v.id) cnt,
                    v.experience
                from vacancies v
                group by v.experience
                order by cnt
             '''
with engine.begin() as conn:
    df_4_5 = pd.read_sql_query(sa.text(query_4_5), conn)
```
